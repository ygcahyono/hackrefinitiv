# Headline News

## Run the queries

### DAX

```python
# list_of_date = []
master_df = pd.DataFrame()
    
for ticker in firm_lists_DAX: 
    
    interval = 17
    init_date = datetime(2022, 1, 1)
    cnt=0
    # ticker = 'BARC.L'

    for rep in range(37):
        to_date = init_date + timedelta(days = interval)

        str_from = datetime.strftime(init_date, '%Y-%m-%d')
        str_to = datetime.strftime(to_date, '%Y-%m-%d')

        temp_df = ek.get_news_headlines( query = f'( Topic:NEWS1 OR Topic:TOPALL OR NS:RTRS) AND R:{ticker} AND Language:LEN'
                              , count=100
                              , date_from = str_from
                              , date_to = str_to )
        
        cnt+=temp_df.shape[0]
        temp_df['firm'] = ticker
        master_df = pd.concat([master_df, temp_df])

        init_date = datetime.strptime(str_to, '%Y-%m-%d')

        if temp_df.shape[0] != 0:
            print(ticker, str_from, '-' , str_to, ':=', temp_df.shape[0], 'news headline')
            
    print('---'*20, f'{ticker} is done with total rows: {cnt}',  '\n')
    
print(f'total number of rows:', master_df.shape[0])
master_df.to_csv(f'../data/output_news_headline_DAX40.csv')
```

```python
# list_of_date = []
master_df = pd.DataFrame()
    
for ticker in firm_lists_DAX: 
        
    try:
        df, errors = ek.get_data(f"{ticker}", ["TR.TRESGCControversiesScoreGrade", "TR.TRESGCControversiesScore", 'TR.TRESGCControversiesScore.date'],
                                 parameters = {"SDate": "0", "EDate":"-10"})

        master_df = pd.concat([master_df, df])
        print(ticker, ': ', df.Date.min()[:10], df.Date.max()[:10])
    except:
        print(ticker, 'does not have ESG controversy score.')
```

```python
master_df[master_df['ESG Controversies Score Grade'] != ''].to_csv('../data/output_ESGControversyScore_DAX.csv')
```
