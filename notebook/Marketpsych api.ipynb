## Function

```python
def Market_psych_data(company, tick):
    
    #presetings for the API
    intro ="https://dataapi.marketpsych.com/rma/v4/data"
    asset_class = "equemea"
    freq = 'dai'
    apikey= 'rf_0niuaILLziXUW8RI0Bg'
    start_date='2022-01-01'
    end_date='2023-05-23'
    format_='csv'
    
    #Search the company to get the asset code
    print(f'looking for {company}...')
    name_search_url = 'https://dataapi.marketpsych.com/rma/v4/equ/assets?apikey=rf_demo12345678&search=' + company
    name_guess_json=requests.get(name_search_url).json()
    name_guess_pd=pd.DataFrame(name_guess_json['data'])
    if len(name_guess_pd)==0:
        print(f'mission value for {company}\n')
    else:
        #we match by Ticker
        if len(name_guess_pd[name_guess_pd['Ticker']==tick])==0:
            print(f'mission value for {company}\n')
        else:
            asset_code=name_guess_pd[name_guess_pd['Ticker']==tick].reset_index()['assetCode'][0]
            print(f'matched to {asset_code} by {tick} \n ')

            #we build the url for the request of sentiment info
            url = f'{intro}/{asset_class}/{freq}/{asset_code}?apikey={apikey}&start_date={start_date}&end_date={end_date}&format={format_}'
            print('making request...')
            #get the request and turn it into a DataFrame
            req=requests.get(url)
            if req.status_code != 200:
                print(f'request failed missing value for {company} \n ')
            else:
                df=pd.read_csv(url)
                print('Done \n ')

                #group by year
                df['Y']=df['windowTimestamp'].str[:4]
                result=pd.pivot_table(df.fillna(0), 
                           index='Y', 
                           columns='dataType', 
                           values= ['laborDispute',
                                  'layoffs',
                                  'managementTrust', 
                                  'innovation',
                                  'mentions',
                                  'buzz'],
                           aggfunc={'laborDispute': np.max, 
                                    'layoffs': np.max, 
                                    'managementTrust': np.max, 
                                    'innovation': np.max,
                                    'mentions': np.sum,
                                    'buzz': np.sum}
                          )
                result['ticker']=tick

                colu=[ (c[0]+'_'+c[1]).replace(' ','') 
                      for c in result.columns.tolist()]
                result.columns=colu

                return result

        
    
```
